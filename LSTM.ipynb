```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
import shap
from tensorflow.keras.optimizers import Adam

# Clear TensorFlow session
tf.keras.backend.clear_session()
print(f"TensorFlow version: {tf.__version__}")

# Load and preprocess data
df = pd.read_csv('air_quality_pollutants_combined_dataset.csv', parse_dates=['date'])
df = df.sort_values('date')

key_pollutants = ['black_carbon', 'nitric_dioxide', 'Ozone', 'PM10', 'PM2.5', 'nitric_oxide']
health_outcomes = ['acute_respiratory_illness']

# Create lagged features (14 lags)
lag_dfs = []
for col in key_pollutants + health_outcomes:
    lags = [df[col].shift(lag) for lag in range(1, 15)]
    lag_df = pd.concat(lags, axis=1, keys=[f'{col}_lag{l}' for l in range(1, 15)])
    lag_dfs.append(lag_df)
X_lags = pd.concat(lag_dfs, axis=1)
df = pd.concat([df, X_lags], axis=1)
df = df.dropna()

original_len = len(pd.read_csv('air_quality_pollutants_combined_dataset.csv'))
print(f"Original rows: {original_len}, Remaining rows: {len(df)}")
print(f"NA counts in first 25 rows: {df.head(25).isna().sum().sum()}")

# Define features and target
X = df[[f'{col}_lag{l}' for col in key_pollutants + health_outcomes for l in range(1, 15)]]
y = df['acute_respiratory_illness']
print(f"X shape: {X.shape}, y length: {len(y)}")

# Normalize features and target
scaler_X = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
scaler_y = StandardScaler()
y_scaled = scaler_y.fit_transform(y.values.reshape(-1, 1)).ravel()

# Create sequences
def create_sequences(X, y, seq_length=14):
    print(f"Creating sequences with X length: {len(X)}, seq_length: {seq_length}")
    if len(X) <= seq_length:
        print(f"Warning: Data length ({len(X)}) is too small. Reducing seq_length to 7.")
        seq_length = 7
    Xs, ys = [], []
    for i in range(len(X) - seq_length + 1):
        Xs.append(X[i:i + seq_length])
        ys.append(y[i + seq_length - 1])
    print(f"Number of sequences created: {len(Xs)}")
    return np.array(Xs), np.array(ys)

seq_length = 14
X_seq, y_seq = create_sequences(X_scaled, y_scaled, seq_length)

# Train-test split
train_size = int(0.8 * len(X_seq))
X_train, X_test = X_seq[:train_size], X_seq[train_size:]
y_train, y_test = y_seq[:train_size], y_seq[train_size:]
print(f"X_train shape: {X_train.shape}, X_test shape: {X_test.shape}")

# Build and train initial LSTM model with adjusted learning rate
model = Sequential([
    tf.keras.Input(shape=(seq_length, X_train.shape[2])),
    LSTM(100, activation='relu', return_sequences=True),
    Dropout(0.3),
    LSTM(50, activation='relu'),
    Dropout(0.3),
    Dense(20, activation='relu'),
    Dense(1)
])
optimizer = Adam(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='mse')
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

print("Training initial model...")
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stopping],
    verbose=1
)

# Evaluate initial model
y_pred_scaled = model.predict(X_test)
y_pred = scaler_y.inverse_transform(y_pred_scaled)
y_test_inv = scaler_y.inverse_transform(y_test.reshape(-1, 1)).ravel()

rmse = np.sqrt(mean_squared_error(y_test_inv, y_pred))
mae = mean_absolute_error(y_test_inv, y_pred)

print("Initial Model Performance:")
print(f"RMSE: {rmse:.4f}")
print(f"MAE:  {mae:.4f}")

# Plot training history
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Training Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout(pad=1.5)
plt.show()

# SHAP Analysis with KernelExplainer (optimized)
n_samples = 68
X_train_flat = X_train[:n_samples].reshape(n_samples, seq_length * X_train.shape[2])
X_test_flat = X_test[:n_samples].reshape(n_samples, seq_length * X_train.shape[2])

def model_predict_reshaped(X_flat):
    n_samples_dynamic = X_flat.shape[0]
    X_seq = X_flat.reshape((n_samples_dynamic, seq_length, X_train.shape[2]))
    return model.predict(X_seq, batch_size=32)

explainer = shap.KernelExplainer(model_predict_reshaped, X_train_flat[:20])
shap_values = explainer.shap_values(X_test_flat, nsamples=200)

shap_values = np.array(shap_values).reshape(n_samples, seq_length, X.shape[1])

# SHAP Summary Plot (Enhanced and Fixed for Scaling)
plt.figure(figsize=(14, 10))  # Increased size for better visibility
max_shap = np.abs(shap_values).max(axis=(0, 1))  # Use max to preserve magnitude
shap.summary_plot(shap_values, X_test[:n_samples], feature_names=X.columns, plot_type="bar", max_display=10, show=False)
bars = plt.bar(range(10), max_shap[np.argsort(max_shap)[::-1][:10]], color=plt.cm.viridis(np.linspace(0, 1, 10)))
plt.xticks(range(10), [X.columns[i] for i in np.argsort(max_shap)[::-1][:10]], rotation=45, ha='right', fontsize=10)
for bar, val in zip(bars, max_shap[np.argsort(max_shap)[::-1][:10]]):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{val:.4f}',
             ha='center', va='bottom' if height > 0 else 'top', fontsize=8)
plt.title('SHAP Feature Importance for LSTM (n=68)', fontsize=14, fontweight='bold')
plt.xlabel('Features', fontsize=12)
plt.ylabel('Maximum Absolute SHAP Value', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.ylim(0, max_shap[np.argsort(max_shap)[::-1][0]] * 1.2)  # Set y-axis limit to 120% of max value
plt.tight_layout(pad=3.0)  # Increased padding
plt.legend().remove()  # Explicitly remove legend if present
plt.show()

# Feature Selection based on SHAP
mean_shap_values = np.abs(shap_values).mean(axis=(0, 1))  # Retain mean for feature selection
feature_importance = pd.Series(mean_shap_values, index=X.columns)
top_features = feature_importance.sort_values(ascending=False).head(20).index.tolist()
print("Top 20 features based on SHAP values:")
print(top_features)

# Save SHAP values
shap_df = pd.DataFrame(shap_values.reshape(n_samples, -1), columns=X.columns.repeat(seq_length))
shap_df.to_csv(f'lstm_shap_values_{n_samples}.csv', index=False)
print(f"SHAP values saved to 'lstm_shap_values_{n_samples}.csv' for {n_samples} samples")

# SHAP Dependence Plots (top pollutants, lag 1)
shap_values_2d = shap_values.mean(axis=1)
X_test_2d = X_test[:n_samples].mean(axis=1)

for feature in [f'{poll}_lag1' for poll in key_pollutants]:
    idx = X.columns.get_loc(feature)
    if idx < X.shape[1]:
        plt.figure(figsize=(8, 6))
        shap.dependence_plot(idx, shap_values_2d, X_test_2d, feature_names=X.columns, show=False)
        plt.title(f'SHAP Dependence Plot for {feature}', fontsize=14, fontweight='bold')
        plt.xlabel('Feature Value', fontsize=12)
        plt.ylabel('SHAP Value', fontsize=12)
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.tight_layout(pad=1.5)
        plt.show()

# Refine Model with Top Features
X_refined = df[top_features]
scaler_X_refined = StandardScaler()
X_scaled_refined = scaler_X_refined.fit_transform(X_refined)

X_seq_refined, y_seq = create_sequences(X_scaled_refined, y_scaled, seq_length)

train_size_refined = int(0.8 * len(X_seq_refined))
X_train_refined, X_test_refined = X_seq_refined[:train_size_refined], X_seq_refined[train_size_refined:]
y_train_refined, y_test_refined = y_seq[:train_size_refined], y_seq[train_size_refined:]

model_refined = Sequential([
    tf.keras.Input(shape=(seq_length, len(top_features))),
    LSTM(100, activation='relu', return_sequences=True),
    Dropout(0.3),
    LSTM(50, activation='relu'),
    Dropout(0.3),
    Dense(20, activation='relu'),
    Dense(1)
])
optimizer = Adam(learning_rate=0.001)
model_refined.compile(optimizer=optimizer, loss='mse')
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

print("Training refined model...")
history_refined = model_refined.fit(
    X_train_refined, y_train_refined,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stopping],
    verbose=1
)

# Evaluate refined model
y_pred_scaled_refined = model_refined.predict(X_test_refined)
y_pred_refined = scaler_y.inverse_transform(y_pred_scaled_refined)
y_test_inv_refined = scaler_y.inverse_transform(y_test_refined.reshape(-1, 1)).ravel()

rmse_refined = np.sqrt(mean_squared_error(y_test_inv_refined, y_pred_refined))
mae_refined = mean_absolute_error(y_test_inv_refined, y_pred_refined)

print("Refined Model Performance:")
print(f"RMSE: {rmse_refined:.4f}")
print(f"MAE:  {mae_refined:.4f}")

# Plot refined training history
plt.figure(figsize=(8, 6))
plt.plot(history_refined.history['loss'], label='Train Loss')
plt.plot(history_refined.history['val_loss'], label='Val Loss')
plt.title('Refined Training Loss Over Epochs', fontsize=14, fontweight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Loss (MSE)', fontsize=12)
plt.legend(loc='upper left')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout(pad=1.5)
plt.show()
```
