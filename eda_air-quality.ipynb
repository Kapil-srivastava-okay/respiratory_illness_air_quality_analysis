### Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datetime import datetime

# Set style for better visualizations
plt.style.use('seaborn-v0_8') 
sns.set_palette("deep")

df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')

# 2. Basic Data Overview
print("=== Dataset Overview ===")
print(f"Shape: {df.shape}")
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
print("\nBasic Statistics:\n", df.describe())
```

```python
# 3. Define health and pollutant columns for analysis
health_columns = ['acute_bronchiolitis_syndromic', 'acute_respiratory_illness', 
                 'influenza_like_syndromic', 'scarlet_fever_syndromic']
pollutant_columns = ['black_carbon', 'blue_particulate_matter', 'green_particulate_matter', 
                     'infra_red_particulate_matter', 'nitric_dioxide', 'nitric_oxide', 
                     'nitrogen_oxides_as_nitrogen_dioxide', 'Ozone', 'PM10', 'PM2.5', 
                     'red_particulate_matter', 'UV_particulate_matter_370nm', 
                     'yellow_particulate_matter_590nm']
```

```python
# 4. Time Series Plots for Health Indicators
import matplotlib.dates as mdates
plt.figure(figsize=(12, 6))

# Plot with clearer lines and smaller markers
for col in health_columns:
    plt.plot(df['date'], df[col], label=col, marker='*', markersize=2, linewidth=1.2)

# Title and labels
plt.title('Time Series of Health Indicators')
plt.xlabel('Date')
plt.ylabel('Cases')

# Legend with larger font
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)

# More date ticks
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=30))  # Adjust interval as needed
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.xticks(rotation=45, fontsize=8)

# Layout adjustments
plt.tight_layout()
plt.show()
```

```python
# 6. Correlation Matrix
plt.figure(figsize=(12, 10))
correlation_matrix = df[health_columns + pollutant_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Health Indicators and Pollutants')
plt.tight_layout()
plt.show()
```

```python
key_pollutants = ['black_carbon', 'nitric_dioxide', 'nitric_oxide', 'nitrogen_oxides_as_nitrogen_dioxide', 'Ozone', 'PM10', 'PM2.5']

# 11. Seasonal Analysis (Monthly Averages)
df['month'] = df['date'].dt.to_period('M')
monthly_avg = df.groupby('month')[health_columns + key_pollutants].mean()

plt.figure(figsize=(12, 6))
for col in health_columns:
    plt.plot(monthly_avg.index.astype(str), monthly_avg[col], label=col, marker='o')
plt.title('Monthly Average of Health Indicators')
plt.xlabel('Month')
plt.ylabel('Average Cases')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 6))
for col in key_pollutants:
    plt.plot(monthly_avg.index.astype(str), monthly_avg[col], label=col, marker='o')
plt.title('Monthly Average of Key Pollutants')
plt.xlabel('Month')
plt.ylabel('Average Concentration (g/m)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Apply 7-day rolling average for health indicators
window_size = 7
for col in health_columns:
    df[f"{col}_rolling"] = df[col].rolling(window=window_size).mean()

# Plot health indicators
plt.figure(figsize=(12, 5), dpi=150)  # Increased width to 14
for col in health_columns:
    plt.plot(df['date'], df[f"{col}_rolling"], label=f"{col} (7-Day Avg)", linewidth=1.5, solid_capstyle='butt')
plt.title("7-Day Rolling Averages of Health Indicators", fontsize=14, fontweight='bold', color='#222')
plt.xlabel("Date", fontsize=12, color='#222')
plt.ylabel("Cases", fontsize=12, color='#222')
plt.xticks(rotation=45, color='#222')
plt.grid(True, linestyle='--', alpha=0.3)  # Subtle grid for clarity
plt.legend(
    bbox_to_anchor=(0.5, -0.15), 
    loc='upper center', 
    fontsize=8, 
    frameon=True, 
    framealpha=0.9, 
    edgecolor='#222', 
    handlelength=1, 
    handleheight=1, 
    borderpad=0.3,
    ncol=len(health_columns)  # Horizontal layout based on number of columns
)
plt.tight_layout()
plt.show()

# Apply 7-day rolling average for pollutants
for col in key_pollutants:
    df[f"{col}_rolling"] = df[col].rolling(window=window_size).mean()

# Plot pollutants
plt.figure(figsize=(12, 5), dpi=150)  # Increased width to 14
for col in key_pollutants:
    plt.plot(df['date'], df[f"{col}_rolling"], label=f"{col} (7-Day Avg)", linewidth=1.5, solid_capstyle='butt')
plt.title("7-Day Rolling Averages of Pollutants", fontsize=14, fontweight='bold', color='#222')
plt.xlabel("Date", fontsize=12, color='#222')
plt.ylabel("Level", fontsize=12, color='#222')
plt.xticks(rotation=45, color='#222')
plt.grid(True, linestyle='--', alpha=0.3)  # Subtle grid for clarity
plt.legend(
    bbox_to_anchor=(0.5, -0.15), 
    loc='upper center', 
    fontsize=8, 
    frameon=True, 
    framealpha=0.9, 
    edgecolor='#222', 
    handlelength=1, 
    handleheight=1, 
    borderpad=0.3,
    ncol=len(key_pollutants)  # Horizontal layout based on number of columns
)
plt.tight_layout()
plt.show()
```

```python
# Seasonal Decomposition of Health Indicators
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# Initialize plot grid: 4 rows x 4 columns
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(20, 12))
fig.suptitle("Seasonal Decomposition of Health Indicators", fontsize=18)

# Loop through health indicators
for row_idx, col in enumerate(health_columns):
    ts = df[col].dropna()

    # Skip if insufficient data
    if len(ts) < 60:
        print(f"Skipping {col} (not enough data points)")
        continue

    # Decompose the time series
    result = seasonal_decompose(ts, model='additive', period=30)

    # Plot into 4 subplots (Observed, Trend, Seasonal, Residual)
    components = [result.observed, result.trend, result.resid]  #result.seasonal
    titles = ['Observed', 'Trend', 'Residual']

    for col_idx, (component, title) in enumerate(zip(components, titles)):
        ax = axes[row_idx, col_idx]
        component.plot(ax=ax, legend=False)
        ax.set_title(f"{col.replace('_', ' ').title()} - {title}")
        ax.set_xlabel("")
        ax.set_ylabel("")

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

### Lag Analysis

```python
# Compute cross-correlations for different lags
max_lag = 30
results = []

for illness in health_columns:
    for pollutant in key_pollutants:
        for lag in range(1, max_lag + 1):
            shifted = df[pollutant].shift(lag)
            corr = df[illness].corr(shifted)
            results.append((illness, pollutant, lag, corr))

# Convert to DataFrame
lag_df = pd.DataFrame(results, columns=['illness', 'pollutant', 'lag', 'correlation'])

# Find top positive correlations (pollution precedes illness)
top_lags = lag_df.sort_values(by='correlation', ascending=False).groupby('illness').head()

top_lags
```

```python
from sklearn.preprocessing import MinMaxScaler

#use it on new dataset dont change the original dataset
# Make a copy of the original dataset
df_new = df.copy()

# Feature Engineering Setup
df_new['date'] = pd.to_datetime(df_new['date'], dayfirst=True)
df_new.set_index('date', inplace=True)

# 1. Create 'season' feature
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Autumn'

df_new['season'] = df_new.index.map(get_season)

# 2. Create 'weekday' and 'is_weekend' features
df_new['weekday'] = df_new.index.day_name()
df_new['is_weekend'] = df_new['weekday'].isin(['Saturday', 'Sunday'])

# 3. Create AQI proxy using normalized key pollutants
scaler = MinMaxScaler()
df_new['aqi_proxy'] = scaler.fit_transform(df_new[key_pollutants]).mean(axis=1)

# Weekday vs. Weekend comparison
weekday_summary = df_new.groupby('is_weekend')[
    key_pollutants + [
        'acute_respiratory_illness',
        'influenza_like_syndromic',
        'acute_bronchiolitis_syndromic',
        'scarlet_fever_syndromic'
    ] + ['aqi_proxy'] 
].mean().round(2)

# Preview updated features and summary
#df_new[['season', 'weekday', 'is_weekend', 'aqi_proxy']].head()
weekday_summary
```

Weekends: Lower traffic-related pollutants, slightly higher illness rates for some conditions

```python
import matplotlib.dates as mdates
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Make sure date is datetime
# If you've already done this earlier, you can skip this line.
df_new['date'] = pd.to_datetime(df_new['date'], format='%d-%m-%Y')

# Style
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))

# Line plot with datetime x-axis
ax = sns.lineplot(data=df_new, x='date', y='aqi_proxy', linewidth=1)

# Format x-axis: Jan, Feb, ...
ax.xaxis.set_major_locator(mdates.MonthLocator())          # monthly ticks
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y')) # e.g., Jan 2024
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')

# Seasonal backgrounds (shade contiguous runs of the same season)
season_colors = {
    'Spring': '#90EE90',
    'Summer': '#FFE4B5',
    'Autumn': '#DEB887',
    'Winter': '#B0E0E6'
}

# Identify contiguous season segments
season_groups = (df_new['season'] != df_new['season'].shift()).cumsum()
legend_added = set()

for _, seg in df_new.groupby(season_groups):
    season = seg['season'].iloc[0]
    start = seg['date'].iloc[0]
    end   = seg['date'].iloc[-1]
    ax.axvspan(start, end, alpha=0.3,
               color=season_colors.get(season, 'gray'),
               label=season if season not in legend_added else None)
    legend_added.add(season)

plt.title('AQI Proxy Over Time with Seasonal Areas')
plt.xlabel('Date')
plt.ylabel('AQI Proxy')
plt.legend(title='Season', ncol=4, frameon=True)
plt.tight_layout()
plt.show()
```
